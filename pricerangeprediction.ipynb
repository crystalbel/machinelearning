```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
data_train = pd.read_csv('../input/technoskill-1/train_techno.csv')
data_train.head()
```

```python
X = data_train.drop('price_range', axis = 1)
y = data_train['price_range']
```

```python
X.head()
```

```python
y.head()
```

```python
data_train.info()
```

```python
data_train.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 24)
```

# Logistic Regression

```python
clf_regression = LogisticRegression().fit(X_train,y_train)
clf_regression.score(X_test,y_test)
```

# Gaussian

```python
gnb = GaussianNB()
clf_gnb = gnb.fit(X_train, y_train)
clf_gnb.score(X_test, y_test)
```

# K-nearest Neighbors

```python
neigh = KNeighborsClassifier(n_neighbors=9)
clf_knn = neigh.fit(X_train, y_train)
clf_knn.score(X_test, y_test)
```

# Decision Tree

```python
clf_dt = tree.DecisionTreeClassifier()
clf_dectree = clf_dt.fit(X_train, y_train)
clf_dectree.score(X_test, y_test)
```

# Random Forest

```python
clf_rf = RandomForestClassifier(max_depth=22)
clf_rf.fit(X_train, y_train)
clf_rf.score(X_test, y_test)
```

# SVC

```python
clf_svc = svm.SVC(kernel = 'poly', gamma = 'auto', verbose = True, degree = 2)
clf_svc.fit(X_train, y_train)
y_predict = clf_svc.predict(X_test)
clf_svc.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_predict))
```

# Confusion Matrix

```python
konfyus = confusion_matrix(y_test, y_predict)
print(konfyus)
```
